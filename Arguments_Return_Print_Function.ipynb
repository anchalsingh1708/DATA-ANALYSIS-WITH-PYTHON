#  *args, **kwargs

```python
#creating function for sum

def total(a,b,c):
    print(a+b+c)

total(1,6,7)
```

```python
#For the above function if value keep increasing every time args will be added 
#hence we will use *args
#*args accepts values in tuple and works only for values of same type like only numbers, string
```

```python
def total(*args):
    print(sum(args))

total(1,8,7,5,4)
```

```python
#If the values are stored in different data types then we will use **kwargs
# **kwargs stores value in key-value pairs(dict)
# works only with keyword arguments, if data is not entered in key and value pairs then gives error
```

```python
def create_user(**kwargs):
    print(type(kwargs))
    print(kwargs)

create_user(name="Anchal",age=23,country="india")

create_user(name="Anchal",last_name="Singh",country="india")
```

```python
def clean_name(name):
    if not name:
        return None
    else:
        
        lo_cleaned=name.strip().lower()
        up_cleaned=name.strip().upper()
        return lo_cleaned,up_cleaned

lo_name,up_name=clean_name("  Anchal  ")

print(lo_name)
print(up_name)
```

```python
# #Python Function habits:
# 1. snake case function name - calculate_discount
# 2. give name to args instead of variable
# 3. Write func info in docstring
# 4. give hints for args  and return value datatype  using : 
```

```python
def caluclate_discount(price:float, rate:float)-> float:
    """
    calculate the final price after applying discount.
    Args:
        price (float): original product price.
        rate (float): discount rate as numbers (eg. 20 for 20%)

    Returns:
        final_price(float): final price after applying discount.

    """

    final_price=price-(price*rate/100)

    return final_price


caluclate_discount(120000, 14)
```

```python
#Transformation functions
#Raw data goes in , gets transformed and comes out processed data

#Task: clean an email and split it into username and domain

def clean_and_split_email(email):
    cl_email=email.strip().lower()

    username , domain=cl_email.split("@")
    return {"username":username,
            "domain": domain}

clean_and_split_email("anchalsingh@gmail.com")
```
