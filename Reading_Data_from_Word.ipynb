```python
import pandas as pd
```

```python
data = [[cell.text for cell in row.cells] for row in table.rows] 
df = pd.DataFrame(data)
```

```python
#spliting data with : as in table values are written as  Name:  Value
```

```python
a=df[0].str.split(':',expand=True)
b=df[1].str.split(':',expand=True)
```

```python
#left of : should be column name and and at right is the record therefore transpose to make it column 
```

```python
a=a.T

b=b.T
```

```python
#combining the 2 data frame which is transposed vertically

data= pd.concat([a,b],axis=1)
```

```python
#header currently is index, making 1st row as header

data=pd.DataFrame(data)
```
