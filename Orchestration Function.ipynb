# Orchestrator Function
#  Controls programm flow by calling other functions in the correct order

```python
#PROJECT

# 1. Receive an email from the user
# 2. Validate the email
# 3. If it is invalid, Log an error in a file.
# 4. If it is valid, clean and structure the email
# 5. Log each step of the program
```

```python
def is_valid_email(email):
    return "@" in email and "." in email

def clean_and_split_email(email):
    cl_email=email.strip().lower()

    username , domain=cl_email.split("@")
    return {"username":username,
            "domain": domain}


def log_write(message):
    with open(r"C:\Users\anchsingh\Anchal Singh\Data Analysis with Python\app.log", "a") as file:
        file.write(message +"\n")
```

```python
#we receive an email from a user
email=input("Please enter your email")

#chek if email is valid
is_valid_email(email)

#if it is not valid, we log the problem
if not is_valid_email(email):
    log_write(f"Invalid Email received: {email}")
else:
    clean_email=clean_and_split_email(email)
    log_write(f"Processed Email: {clean_email}")
log_write("App Stopped")
```

# Orchestration Function

```python

def process_user_email(email):
    #check if email is valid
    is_valid_email(email)
    
    #if it is not valid, we log the problem
    if not is_valid_email(email):
        log_write(f"Invalid Email received: {email}")
    #if it is valid, we clean it and store structured information
    else:
        clean_email=clean_and_split_email(email)
        log_write(f"Processed Email: {clean_email}")
    log_write("App Stopped")


#we receive an email from a user
email=input("Please enter your email")
process_user_email(email)
```
